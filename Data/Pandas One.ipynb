```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Salaries.csv').replace('Not Provided', np.nan).astype({"BasePay": float, "OtherPay": float})
```

```python
df.head(10)
```

```python
pd.set_option('max_colwidth', 40)
pd.set_option('precision', 4)
pd.set_option('chop_threshold', .5)
```

```python
df.head(10).style.format({
    "BasePay": "${:20,.0f}",
    "OtherPay": "${:20,.0f}",
    "TotalPay": "${:20,.0f}",
    "TotalPayBenefits": "${:20,.0f}",
    "JobTitle": lambda x: x.title(),
    "EmployeeName": lambda x: x.title()
}).hide_index() \
.highlight_max(color='lightgreen') \
.highlight_min(color='#cd4f39')
```

```python
df.head(10).style.format({"BasePay": "${:20,.0f}", 
                          "OtherPay": "${:20,.0f}", 
                          "TotalPay": "${:20,.0f}",
                          "TotalPayBenefits":"${:20,.0f}"})\
                 .format({"JobTitle": lambda x:x.lower(),
                          "EmployeeName": lambda x:x.lower()})\
                 .hide_index()\
                 .background_gradient(cmap='Blues')
```

```python
html = df.head(10).style.format({"BasePay": "${:20,.0f}", 
                          "OtherPay": "${:20,.0f}", 
                          "TotalPay": "${:20,.0f}",
                          "TotalPayBenefits":"${:20,.0f}"})\
                 .format({"JobTitle": lambda x:x.lower(),
                          "EmployeeName": lambda x:x.lower()}).hide_index().background_gradient(cmap='Blues').render()
```

```python
css = """<style>table, th, td {
  border: 1px solid black;
}</style>"""

final_html = """<html><head></head>{1}<div>{0}</div></html>""".format(html, css)
```

```python
with open('one.html', 'w+') as html_file:
    html_file.writelines(final_html)
```
